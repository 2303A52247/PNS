# Question 1: Generate Normal Distribution and Plot Histogram


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

mean_1 = 60
std_dev_1 = 12
size_1 = 1000

data_1 = np.random.normal(mean_1, std_dev_1, size_1)

plt.figure(figsize=(8, 5))
plt.hist(data_1, bins=30, density=True, alpha=0.6, color='blue', edgecolor='black')

x = np.linspace(mean_1 - 4*std_dev_1, mean_1 + 4*std_dev_1, 1000)
pdf = stats.norm.pdf(x, mean_1, std_dev_1)
plt.plot(x, pdf, color='red', linewidth=2, label='PDF')

plt.title('Histogram and PDF of Normal Distribution')
plt.xlabel('Data Points')
plt.ylabel('Density')
plt.legend()
plt.show()

```

# Question 2: Calorie Intake Probabilities

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

mean_2 = 30
std_dev_2 = 5

prob_greater_19 = 1 - stats.norm.cdf(19, mean_2, std_dev_2)
prob_less_19 = stats.norm.cdf(19, mean_2, std_dev_2)
prob_between_19_35 = stats.norm.cdf(35, mean_2, std_dev_2) - stats.norm.cdf(19, mean_2, std_dev_2)

print(f"Probability of daily calorie intake >= 19: {prob_greater_19:.4f}")
print(f"Probability of daily calorie intake <= 19: {prob_less_19:.4f}")
print(f"Probability of daily calorie intake between 19 and 35: {prob_between_19_35:.4f}")

x = np.linspace(mean_2 - 4*std_dev_2, mean_2 + 4*std_dev_2, 1000)
pdf = stats.norm.pdf(x, mean_2, std_dev_2)

plt.figure(figsize=(8, 5))
plt.plot(x, pdf, color='blue', linewidth=2, label='PDF')
plt.fill_between(x, pdf, where=(x >= 19), color='red', alpha=0.3, label='>= 19')
plt.fill_between(x, pdf, where=(x <= 19), color='green', alpha=0.3, label='<= 19')
plt.fill_between(x, pdf, where=((x >= 19) & (x <= 35)), color='purple', alpha=0.3, label='19 to 35')

plt.title('Probability Distribution of Daily Calorie Intake')
plt.xlabel('Calories')
plt.ylabel('Density')
plt.legend()
plt.show()
```
